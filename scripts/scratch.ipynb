```python
import pandas as pd
import yfinance as yf
```

```python
portfolio = pd.read_csv(r"../data/portfolio.csv")
portfolio.head()
```

```python
port_tickers = list(portfolio.symbol)

df = yf.download(port_tickers, period='1mo',auto_adjust=True)
df.head()
```

```python
df.to_csv(r'../data/port_ticker_history_1mo.csv')
```

```python
adj_close_df = df['Close']
adj_close_df
adj_close_df.to_csv(r'../data/port_ticker_history_1mo_adj_close.csv')
```
